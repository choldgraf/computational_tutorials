```python
mysin = sin(np.arange(0, 8*np.pi, .01))
linamp = np.linspace(0, 5, mysin.shape[0])
corramp = make_correlated_noise(mysin.shape[0], stdev=1)
snd = pd.DataFrame(mysin * corramp.values.squeeze())
noise = 3*sp.random.randn(mysin.shape[0])

w = 2
y = pd.DataFrame(w * corramp.values.squeeze() + noise)
```
