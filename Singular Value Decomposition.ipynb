This script simulates SVD and Cholesky Decomposition
to show you what it's doing

```python
import matplotlib as mpl
import scipy.io as si
```

<h3>First simulate some data</h3>

```python
covA = np.array([[2, 1],[1, 2]]) # covariance matrix for the data
aUnWhite = np.linalg.cholesky(covA)
randA = np.random.randn(1000, 2)
a = dot(randA, aUnWhite) # this gives it the covariance we want
```

### First visualize what it should end up looking like

```python
fig = plt.figure(1, figsize=(10, 5))
subplot(121)
plt.plot(randA[:, 0], randA[:, 1], '.')
plt.title('Data before covariance')

subplot(122)
plt.plot(a[:, 0], a[:, 1], '.')
axLim = a.max() * 1.2
plt.axis([-axLim, axLim, -axLim, axLim])
plt.title('Now w/ covariance applied')
```

Now we will actually do the SVD:

\begin{align}
A = U\Sigma V'
\end{align}

```python
[left, mid, right] = np.linalg.svd(a, full_matrices=False)
mid = np.diag(mid)
```

### Now we can see what the matrices we get from SVD actually gives us

```python
# Plot the left singular vectors
figure(2, figsize(15, 5))
subplot(1, 3, 1)
axLim2 = left.max() * 1.2
plt.plot(left[:, 0], left[:, 1], '.')
plt.axis([-axLim2, axLim2, -axLim2, axLim2])
plt.title('Original cloud (left matrix)')

# Now scale
scaled = np.dot(left, mid)
subplot(1, 3, 2)
plt.plot(scaled[:, 0], scaled[:, 1], '.')
plt.axis([-axLim, axLim, -axLim, axLim])
plt.title('Now scale them')

# Now rotate back
orig = np.dot(scaled, right)
subplot(1, 3, 3)
plot(orig[:, 0], orig[:, 1], '.')
axis([-axLim, axLim, -axLim, axLim])
plt.title('Now rotate them')
```

### Now compare to the original datapoints

```python
plt.figure(3, figsize(10, 5))
# Original Data
subplot(121)
plt.plot(a[:, 0], a[:, 1], '.')
axLim = a.max() * 1.2
plt.axis([-axLim, axLim, -axLim, axLim])
plt.title('Original Data')

# SVD Data
subplot(122)
plot(orig[:, 0], orig[:, 1], '.')
axis([-axLim, axLim, -axLim, axLim])
plt.title('After SVD')
```

## Cholesky Decomposition

The Cholesky decomposition is basically a way of taking the "square root" of a matrix.  It is a matrix decomposition which states that a squre matrix can be constructed as the inner product of a triangular matrix with itself.  This is particularly useful in the case of covariance matrices, which are always square.

\begin{align}
A=U^TU
\end{align}

So, first we'll simulate some data and see what this looks like.  Note that in constructing the data itself, we can use the Cholesky decomposition.  More on that later.

```python
# import some matlab data
myData = '/Users/choldgraf/Desktop/wordChoiceTempStorage/spectrograms_ID_Kira_snr_0.0001.mat'
matData = si.loadmat(myData)
mySpecs = matData['specMatrix']
```

```python
vizSpec = mySpecs[8, :, :]
imshow(vizSpec, vmax=.5*vizSpec.max())
```

```python
covViz = cov(vizSpec)
diagMatrix = np.diag([.01*covViz.max()]*covViz.shape[0]) # Adding some diagonal elements so that it's positive semi-definite
covViz = covViz + diagMatrix
covVizColMax = .2*covViz.max()
imshow(covViz, vmax=covVizColMax)
```

Now we'll apply the Cholesky decomposition to this and see what it looks like.  Note that it is now a lower triangular matrix.

```python
cholViz = np.linalg.cholesky(covViz)
imshow(cholViz)
```

```python
subplot(121)
imshow(covViz, vmax=covVizColMax)
subplot(122)
imshow(dot(cholViz.T, cholViz), vmax=covVizColMax)
```
